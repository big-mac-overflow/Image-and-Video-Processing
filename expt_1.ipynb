Image Blending

```python
# importing library for resizing(sh)
from skimage.transform import resize

# division by 255 done to normalize values(otherwise the values add up upon superimposition and we get white image)
img1=io.imread('dhoni.jpg')/255
img2=io.imread('ViratKohli.jpg')/255

sh=img2.shape
img1=resize(img1,(sh[0],sh[1],3)) # resize img1 to the shape of image two, otherwise we get error 'ValueError: operands could not be broadcast together with shapes (258,196,3) (200,200,3)' which simply means only images that are the same size, can be superimposed
                                  # sh[0] is index 0, sh[1] is index 1
alpha = 0.5
imgr = ((1-alpha)*img1) + alpha*(img2) #imgr is resultant image
io.imshow(imgr)
```
