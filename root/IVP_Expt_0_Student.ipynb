Change image to darker color

```python
img=io.imread('cat.png')

plt.subplot(121) # 1st row, 1st colums
io.imshow(img)
plt.title('Original image')

im1=img
im1[img<(0.75*255)]=0 # change this(play with it)


plt.subplot(122) # 1st row, 2nd colums
plt.imshow(im1)
plt.title('Dark image')
```
