# Low Pass filtering for blurring

It is also known as smoothing filter. It removes the high frequency content from the image. It is also used to blurr an image. Low pass averaging filter mask is as shown.<br>

1/9 1/9 1/9<br>
1/9 1/9 1/9<br>
1/9 1/9 1/9<br>

the pixel which is to be filtered is processed with mask as shown in example below<br>
The pixel with value 106 after applying mask is converted to 115.<br>
The mask is shifted pixel by pixel and the process is repeated till all pixels are covered.


image.png

````python
#import io, numpy, matplotlib
from skimage import io
import numpy as np
import matplotlib.pyplot as plt

# importing for rgb2gray conversion
from skimage.color import rgb2gray

# the following library includes convolution
from scipy import signal
````

````python
# reading image
img = io.imread('boat.png')

# creating mask
# bigger the mask, more the blurring

# # 3 mask
# mask = np.ones([3,3], dtype = int)
# mask = mask / 9

# #7 mask
# mask = np.ones([7,7], dtype = int)
# mask = mask / 49

# 11 mask
mask = np.ones([11,11], dtype = int)
mask = mask / 121

# Applying convolution 
avg_filt = signal.convolve2d(img, mask)

# Plotting images
plt.figure(figsize =(30,30))

plt.subplot(121)
plt.imshow(img, cmap = 'gray')
plt.title('Original Image')

plt.subplot(122)
plt.imshow(avg_filt, cmap = 'gray')
plt.title('Image after average filtering')
````

````python
# reading image
img = io.imread('boat.png')

# creating mask
mask = np.ones([3,3], dtype = int)

# Laplacian mask
# mask = np.ones([5,5], dtype = int)
mask = np.array( [[0, 1, 0] , [1, -4, 1] , [0, 1, 0]] ) 
mask

# Apply convolution between image and mask 
output = signal.convolve2d(img, mask)

# Plotting the original and filtered images
plt.figure(figsize =(30,30))

plt.subplot(121)
plt.imshow(img, cmap = 'gray')
plt.title('Original Image')

plt.subplot(122)
plt.imshow(output, cmap = 'gray')
plt.title('Image after laplacian filtering')
````

````python
# same but for Blurr_moon.tif
# plot the original and resultant image

# reading image
img = io.imread('Blurr_moon.tif')

# creating mask
mask = np.ones([3,3], dtype = int)

# Laplacian mask
# mask = np.ones([5,5], dtype = int)
mask = np.array( [[0, 1, 0] , [1, -4, 1] , [0, 1, 0]] ) 
mask

# Apply convolution between image and mask 
output = signal.convolve2d(img, mask)

# Plotting the original and filtered images
plt.figure(figsize =(30,30))

plt.subplot(121)
plt.imshow(img, cmap = 'gray')
plt.title('Original Image')

plt.subplot(122)
plt.imshow(output, cmap = 'gray')
plt.title('Image after laplacian filtering')


````
